가격 예측

타임시리즈 데이터를 예측하는 라이브러리 중에
페이스북이 개발한 Prophet 라이브러리를 사용

프로펫은 2개 컬럼만 있으면, 예측 가능하다
날짜 컬럼과, 예측하려는 컬럼


```python
df_avocado =df[['Date', 'AveragePrice']]
```

```python
df_avocado
```

컬럼명을, 프로펫 라이브러리가 인식할 수 있도록 바꿔준다.

```python
df_prophet =df_avocado.rename(columns = {'Date':'ds','AveragePrice':'y' })
```

```python
df_prophet
```

2. 데이터를 학습시킨다.

```python
from prophet import Prophet

m = Prophet()
```

```python
m.fit(df_prophet)
```

3. 예측에 사용할 데이터 프레임을 만든다.

```python
future =m.make_future_dataframe(periods = 365, freq='D')
```

4. 예측한다.

```python
forecast=  m.predict(future)
```

5. 예측한 결과를 차트로 나타낸다.

```python
m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
plt.savefig('fig1.jpg') # 그림파일을 jpg 파일로 저장
```

```python
m.plot_components(forecast)
plt.savefig('fig1.jpg')
```

지금까지의 예측은. 미국 지역 전체에 대한 데이터 예측이었다

```python
df_west =df.loc[df['region'] == 'West', ]
```

```python
df_west
```

```python
df_prophet = df_west[['Date', 'AveragePrice']]
```

```python
df_prophet.rename(columns={'Date':'ds', 'AveragePrice':'y'}, inplace=True)
```

```python
df_prophet
```

```python
m2 = Prophet() 
```

```python
m2.fit(df_prophet)
```

```python
future2 = m2.make_future_dataframe(periods=365, freq='D')
```

```python
forecast2 = m2.predict(future2)
```

```python
m2.plot(forecast2)
plt.savefig('fig1.jpg')
```

```python
m2.plot_components(forecast2)
plt.savefig('fig1.jpg')
```
